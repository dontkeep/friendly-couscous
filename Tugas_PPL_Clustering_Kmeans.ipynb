import pandas untuk melakukan transformasi data sebelum dilakukan clustering

```python
import pandas as pd
```

```python

# Get the file path of the dataset on Google Drive
file_path = '/content/drive/MyDrive/ppl/Mall_Customers.csv'

# Read the dataset into a Pandas DataFrame
df = pd.read_csv(file_path)
df
```

menghapus column Genre karena berjenis object sedangkan untuk melakukan cluster sebaiknya berjenis int atau float.

```python
df = df.drop(columns=['Genre', 'CustomerID', 'Age'])
df
```

membuat file baru dengan dataset yang sesuai untuk dilakukan clustering

```python
# prompt: create new csv file using the new df

df.to_csv('new_mall_customers.csv', index=False)
```

visualisasi hasil clustering dari program c++

After clustering

```python
import matplotlib.pyplot as plt
import seaborn as sns
# After clustering
output_path = '/content/drive/MyDrive/ppl/output.csv'
plt.figure()
df = pd.read_csv(output_path)
sns.scatterplot(x=df.x, y=df.y,
                hue=df.c,
                palette=sns.color_palette("hls", n_colors=5))
plt.xlabel("Annual income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.title("Clustered: spending (y) vs income (x)")

plt.show()
```
